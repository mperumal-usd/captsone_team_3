# Spotify Basic Pitch Test

```python
from basic_pitch.inference import predict
from basic_pitch import ICASSP_2022_MODEL_PATH
from pathlib import Path
import pretty_midi
```

```python
def load_midi(midi_path):
    try:
        return pretty_midi.PrettyMIDI(midi_path)
    except Exception as e:
        #print(f"⚠️ Skipping {midi_path}: {e}")
        return None
def test_midi_file(file):
    tmpOpen = load_midi(file)
    if tmpOpen == None:
        return False
    return True
```

```python
# Setting file system variables
project_directory = f"C:\\Users\\isral\\MSAAI\\AAI_590"
midi_directory = f"{project_directory}\\captsone_team_3\\MidiDatasets\\590-Classical-music-midi"
excluded_directory = f"{project_directory}\\captsone_team_3\\MidiDatasets\\InvalidFiles"

print("Project directory:", project_directory)
print("MIDI directory:", midi_directory)
print("Exists:", Path(midi_directory).exists())
```

```python
from pathlib import Path
import shutil

# midi_directory and excluded_directory are assumed to be Path objects or strings
midi_directory = Path(midi_directory)
excluded_directory = Path(excluded_directory)

# Ensure excluded directory exists
excluded_directory.mkdir(parents=True, exist_ok=True)

total_files = 0
excluded_files = 0
for file in midi_directory.rglob("*"):
    # Only process files that look like MIDI
    if file.is_file() and file.suffix.lower() in {".mid", ".midi"}:
        total_files += 1
        # If your test function returns False, move the file
        if not test_midi_file(file):
            # Compute the relative path from the root midi_directory
            relative_path = file.relative_to(midi_directory)

            # Recreate the same subfolder structure inside excluded_directory
            destination = excluded_directory / relative_path
            destination.parent.mkdir(parents=True, exist_ok=True)

            shutil.move(str(file), str(destination))
            print(f"Excluded: {relative_path}")

            # Track the number of excluded files
            excluded_files += 1

print(f'Total MIDI files found: {total_files}. Total excluded files: {excluded_files}')
```

```python


audio_file = f"C:\\Users\\isral\\MSAAI\\AAI_590\\AudioFiles\\Something-Voice-Verse1-2.wav"

#Loading the Note events, midi data, and notes list from the prediction
note_events, midi_data, note_list = predict(
    audio_path=audio_file,
    onset_threshold=0.5,
    frame_threshold=0.3,
    minimum_note_length=64,
    multiple_pitch_bends=False,
    midi_tempo=73,
    model_or_model_path=ICASSP_2022_MODEL_PATH
)
# Remove all pitch bends
for instrument in midi_data.instruments:
    instrument.pitch_bends = []

#Next, saving the output midi file:
output_file = f"C:\\Users\\isral\\MSAAI\\AAI_590\\MidiFiles\\Something-Voice-Verse1-2-0_5-0_3-64-NoPitchBend-73.mid"

midi_data.write(output_file)
print("Saved:", output_file)
```
